```python
def heuristicStep(n, m, g, subset):
    maxCount = -1
    maxRow = -1
    for row in range(len(g)):
        if not(row in subset):
            #counts = g[row].count(1)
            counts = np.count_nonzero(g[row])
            if maxCount < counts:
                maxCount = counts
                maxRow = row
    return maxRow 
        
def runGreedy(n, m, g):
    subset = []
    iters = -1
    while (n>0) and (iters < n):
        row = heuristicStep(n,m,g,subset)
        subset.append(row)
        colsThatRemain = [] # only keep reads that have not been accounted for
        for i in range(n):
            if g[row][i] == 0:
                colsThatRemain.append(i)
        g_new = [] 
        for r in g:
            rNew = []
            for col in colsThatRemain:
                rNew.append(r[col])
            g_new.append(rNew)
        if len(g_new) != 0:
            n = len(g_new[0])
        g = g_new
        iters += 1
    if n==0:
        return subset
    else:
        return []

#print(runGreedy(2,2,[[0,0],[1,0]]))
```
